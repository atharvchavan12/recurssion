```python
#2. Given a recurrence relation, solve it using the substitution method:

#a. T(n) = T(n-1) + c

#b. T(n) = 2T(n/2) + n

#c. T(n) = 2T(n/2) + c

#d. T(n) = T(n/2) + c

"""
a. T(n) = T(n-1) + c

We can directly substitute T(n-1) with its equivalent expression and continue until we reach the base case.

T(n) = T(n-1) + c
= (T(n-2) + c) + c
= T(n-2) + 2c
= (T(n-3) + c) + 2c
= T(n-3) + 3c
...
= T(n-k) + kc

If we let n-k = 0, then k = n.

Therefore, T(n) = T(0) + nc.

b. T(n) = 2T(n/2) + n

We'll substitute T(n/2) repeatedly, halving n each time:

T(n) = 2T(n/2) + n
= 2(2T(n/4) + n/2) + n
= 4T(n/4) + 2n
= 4(2T(n/8) + n/4) + 2n
= 8T(n/8) + 3n
...
= 2^kT(n/2^k) + kn

If we let n/2^k = 1, then k = log2(n).

Therefore, T(n) = 2^log2(n)T(1) + nlog2(n) = nT(1) + nlog2(n).

c. T(n) = 2T(n/2) + c

Similar to the previous case, we substitute T(n/2) repeatedly:

T(n) = 2T(n/2) + c
= 2(2T(n/4) + c) + c
= 4T(n/4) + 3c
= 4(2T(n/8) + c) + 3c
= 8T(n/8) + 7c
...
= 2^kT(n/2^k) + (2^k - 1)c

Again, letting n/2^k = 1, we get k = log2(n).

Therefore, T(n) = 2^log2(n)T(1) + (2^log2(n) - 1)c = nT(1) + (n - 1)c.

d. T(n) = T(n/2) + c

Substituting T(n/2) repeatedly:

T(n) = T(n/2) + c
= (T(n/4) + c) + c
= T(n/4) + 2c
= (T(n/8) + c) + 2c
= T(n/8) + 3c
...
= T(n/2^k) + kc

Letting n/2^k = 1, we get k = log2(n).

Therefore, T(n) = T(1) + log2(n)c.

By using the substitution method, we've derived the closed-form solutions for each recurrence relation.
"""
```
